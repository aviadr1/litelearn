```python
%load_ext autoreload
%autoreload 2

import seaborn as sns
import litelearn as ll
```

```python
dataset = "penguins"
target = "species"

df = sns.load_dataset(dataset)
df = df.dropna(subset=[target])
model = ll.classify_df(
    df,
    target,
    use_nulls=True,
)
pred = model.predict(df)
# model.display_evaluation()
```

```python
model.display_evaluation()
```

```python
from sklearn.metrics import classification_report
import pandas as pd
# pd.DataFrame(classification_report(df[target], pred, target_names=model.train_frame.y_train.cat.categories, output_dict=True))
classification_report(df[target], pred, target_names=model.train_frame.y_train.cat.categories, output_dict=True)
```

```python
report = classification_report(df[target], pred, target_names=model.train_frame.y_train.cat.categories, output_dict=True)
    
report.update({"accuracy": {"precision": None, "recall": None, "f1-score": report["accuracy"], "support": report['macro avg']['support']}})
ans = pd.DataFrame.from_dict(report).T
ans
```
